```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel('varejo.xlsx')
vendas.head()
```

```python
vendas.info()
```

```python
vendas.groupby('idcanalvenda').idcompra.nunique() # Agrupando e contando quantas vendas tem por cada canal de vendas
```

```python
vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace('APP','Aplicativo') # Mudando na coluna "idcanalvendas" o nome de "APP"
vendas.groupby('idcanalvenda').idcompra.nunique()                               # para "Aplicativo" pois são a mesma coisa
```

```python
vendas.groupby('bandeira').idcompra.nunique()
```

```python
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas.groupby('estado').idcompra.nunique()
```

```python
vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace(' ','_') # Trocando todos os espaços em branco entre os nomes
vendas.groupby('Nome_Departamento').idcompra.nunique()                         # e trocando por underscores(_)
```

```python
vendas[vendas['estado'].isnull()] # Função que retorna todos os valores nulos da coluna "estado"
                                  # No database de vendas
```

```python
vendas_sem_nulos = vendas.dropna() # Função que Exclui todos os valores de todas as linhas
                                   # que possuem dados nulos
vendas_sem_nulos.info()
```

```python
vendas['estado'].fillna('MS',inplace=True) # .fillna() função que troca todos os nulos na coluna estado pela
                                           # sigla "MS"(MAto Grosso do Sul) "inplace = True" significa que
                                           # se quer manter a alteração no dataset original sem a necessidade
                                           # de atribuição.
```

```python
vendas.info()
```

```python
media_preço = vendas['Preço'].mean()             # Função que tira a media de preços do dataset
vendas['Preço'].fillna(media_preço,inplace=True) # Função que coloca a media de preços no lugar dos nulos na coluna "Preços"
```

```python
vendas.info()
```

```python
vendas.head()
```

```python
vendas[['Preço','Preço_com_frete']].describe()
```

```python
vendas_Correto = vendas.query('Preço < Preço_com_frete') # Atribuindo a uma nova variavel o dataset sem
vendas_Correto.describe()                                # o erro de que o Preço_com_frete é menor que o
                                                         # Preço
                                                         # Vendas_Correto sera o novo dataset
```

```python
vendas_Correto.head()
```

```python
# Respondendo a primeira metrica, Quais ops departamentos mais vendidos
# Pegando os departamentos com mais vendas e agrupando do meior para o menor

vendas_Correto.groupby('Nome_Departamento').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
# Respondendo a segunda metrica. Média de preço com frete por departamento
# Agrupando os departamentos pelo Preço com frete e tirando a media
# Depois prganizando do maior para o menor e printando como uma tabela
# Round() função que coloca quantas casas decimais voce quer mostrar

round(vendas_Correto.groupby('Nome_Departamento').Preço_com_frete.agg('mean').sort_values(ascending=False).reset_index(),2)
```

```python
# Respondendo a terceira metrica. Quantidade de vendas por data
vendas_Correto.groupby('Data').idcompra.nunique().sort_values(ascending=False)
```

```python
vendas_Correto['mes'] = vendas_Correto['Data'].dt.month # Criando uma coluna nova no dataset
                                                        # e colocando nela o mes em que cada compra foi feita
                                                        # .dt = Fução que pega a data da coluna
                                                        # .month = Pega o mes da coluna
vendas_Correto.head()
```

```python
# Respondendo a terceira metrica. Quantidade de vendas por mes
vendas_Correto.groupby('mes').idcompra.nunique().sort_values(ascending=False)
```

```python
cliente = pd.read_excel('/content/cliente_varejo.xlsx') # Importando o segundo dataset
cliente.head()
```

```python
cliente[['idade','renda']].describe()
```

```python
cliente.info()
```

```python
cliente = cliente.astype({'renda':'float'}) #transformando a coluna "renda" em Float ao inves de int
cliente.info()
```

```python
# Como a coluna 'cliente_log' é comum entre os dois datasets, Pode-se unir os dois, por meio de joins
#Left join
vendas_cliente = vendas_Correto.merge(cliente,how='left',on='cliente_Log') # Atribuindo a nova tabela a "Venda_correto" como tabela principal a tabela "Cliente"
                                                                           # .merge(1,2,3) = Função que possibilida o Join
                                                                           # 1 - Parametro que voce pega a tabela secundaria para juntar com a principal
                                                                           # 2 - "how=" Como voce quer fazer esta junção (Left é o mais comum)
                                                                           # 3 - "on=" Por qual coluna em comum sera essa junção dos dois datasets
vendas_cliente.head()
```

```python
# Respondendo a Quarta Metrica. Media da renda para cada tipo de canal de venda

renda_canalvenda = round(vendas_cliente.groupby('idcanalvenda').renda.agg('mean').sort_values().reset_index(),2)
renda_canalvenda
```

```python
# Respondendo a Quinta MEtrica. Media de idade de clientes por bandeira

idade_bandeira = round(vendas_cliente.groupby('bandeira').idade.agg('mean').reset_index(),2)
```

```python
# Importando Bibliotecas de vizualizações de dados
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(7,4)) # Atribuindo a "fig" o tamanho em que a figura sera plotada
                                 # plt.figure(1) = função que configura o tamanho da figura
                                 # 1 - figsize() = definindo o tamnho da figura

plt.bar(idade_bandeira['bandeira'],idade_bandeira['idade'],color='green') # plt.bar(1,2,3) = função que plota um grafico de barras
                                                                          # 1 - Indicando qual informação do data set sera o eixo X do grafico de barras
                                                                          # 2 - Indicando qual informação do dataset sera o eixo Y do grafico de barras
                                                                          # 3 - Indicando a cor do grafico de barras

plt.ylabel('Media de idade')  # plt.ylabel() = Colocando o nome no eixo Y

plt.title('Idade media por bandeira',loc='center',fontsize=18,fontweight='bold',color='black'); # plt.title(1,2,3,4,5) = Função que coloca um titulo no grafico de barros
                                                                                                # 1 - Nome do Grafico
                                                                                                # 2 - "loc=" Localização em que o nome ficara
                                                                                                # 3 - "fonsize=" Tamanho da fonte do titulo
                                                                                                # 4 - "fontweight=" Estilo da fonte usado do Titulo
                                                                                                # 5 - "color=" Cor da fonte do titulo
```

```python
fig2 = plt.figure(figsize=(7,4))

plt.bar(renda_canalvenda['idcanalvenda'],renda_canalvenda['renda'],color='blue')

plt.ylabel('Renda')

plt.title('Renda média por canal de venda',loc='center',fontsize=18,fontweight='bold',color='black');
```

```python
vendas_por_data = vendas_Correto.groupby('Data').idcompra.nunique().reset_index()

vendas_por_data
```
