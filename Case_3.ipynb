```python
base_casa = base_casa.rename(columns={'Avg. Area Income':'Avg_Area_Income','Avg. Area House Age':'Avg_Area_House_Age','Avg. Area Number of Rooms':'Avg_Area_Number _of_Rooms',
                                      'Avg. Area Number of Bedrooms':'Avg_Area_Number_of_Bedrooms','Area Population':'Area_Population'})
base_casa.columns
```

```python
# Tirando a coluna de endereço do dataset, pois as informações nela contidas não seram necessarias para o desenvolvimento da regressão Linear

base_casa.drop(["Address"],axis=1,inplace=True) # Eliminando a coluna "Address" do dataset
                                                # "Axis=1" - Elminar a coluna. "Axis=0" elimina a linha
                                                # "inplace-True" - Para modificar em cima do dataset original
base_casa.head()
```

```python
x = base_casa[['Avg_Area_Income', 'Avg_Area_House_Age', 'Avg_Area_Number _of_Rooms', #Definindo as variaveis explicattivas
       'Avg_Area_Number_of_Bedrooms', 'Area_Population']]                            # Variaveis X

y = base_casa['Price']                                                               # Definindo a varivel Y (Variavel resposta)
```

```python
from sklearn.model_selection import train_test_split    #Importando uma biblioteca voltada para Machine lerning
                                                        # No qual divide-se o dataset em duas partes. A parte de treinamento do machine lerning
                                                        # E a parte de teste.
                                                        # Se os resultados dos dois forem parecidos ou iguais, o modelo esta preparado para o tipo de analise


x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.7, test_size=0.3, random_state=42) #Criando quatro variaveis. Duas para treinamento (x_train e y_train)
                                                                                                         # Duas para teste (x_test e y_test)
                                                                                                         # train_test_split(1,2,3,4,5)
                                                                                                         # 1 - Passando o valor de x definido anteriormente
                                                                                                         # 2 - Passando o valor de y definido anteriormente
                                                                                                         # 3 - "Train_size=" Definindo o tamanho da base de treino
                                                                                                         # Usa-se em media de 70% a 80% da base de dados
                                                                                                         # 4 - "test_size=" Definindo o tamanho da base de teste
                                                                                                         # 5 -  "random_state=" para pegar dados aleatorios
```

```python
print(x_train.shape) #Printando o tamanho de cada base
print(x_test.shape)  #.shape mostra qauntos dados tem dentro da base
print(y_test.shape)
print(y_train.shape)
```

```python
from sklearn.linear_model import LinearRegression  #importando uma biblioteca expecifica para uma regressão linear
```

```python
lm = LinearRegression() # Atribuindo a função LinearRegression a variavel lm
```

```python
lm.fit(x_train,y_train) #Treinando a base de dados
```

```python
y_pred = lm.predict(x_test) #.predict() vai predizer o valor com base no treinamento
```

```python
from sklearn.metrics import r2_score # Importando o "sklearn.metrics" para se ter as metricas do sklearn
                                     # Importando o "r2_score". Pois quanto maior o R2 Score maior é a possibilidade de se ter um acerto.
                                     # Maior é a possibilidade de se conseguir a regressão linear

r = r2_score(y_test,y_pred)          #"r2_score" irá comparar o "y_pred" que é a predição e o "y_test" que é o teste que foi feito
                                     # e o resultado desta comparação sera atribuido a variavel "r"
```

```python
print('r_Quadrado:',r)    # O resultado de "r_Quadrado: 0.914681849875402"
                          # Significa que se tem 91% de chance de quando as variveis explicativas mudarem
                          # a Variavel y (target) também mudará.
                          # Significa que se tem uma linearidade muito alta dentro deste modelo
                          # Os Resultados vao ser bem preditos
```
